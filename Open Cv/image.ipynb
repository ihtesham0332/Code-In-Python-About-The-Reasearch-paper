Make the Image binary formate

```python

import cv2

# Step 1: Read the image
image = cv2.imread('1.jpg')
resiz=cv2.resize(image,(300,300))
# Step 2: Convert the image to grayscale
gray_image = cv2.cvtColor(resiz, cv2.COLOR_BGR2GRAY)

# Step 3: Apply thresholding to segment the image
_, segmented_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)

# Step 4: Display the segmented image
cv2.imshow('Segmented Image', segmented_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Mean Shifting Formula Using

```python
import cv2
import matplotlib.pyplot as plt

# Step 1: Read the image
image = cv2.imread('1.jpg')

# Step 2: Perform Mean-Shift segmentation
mean_shifted_image = cv2.pyrMeanShiftFiltering(image, 10, 30)  # Adjust the parameters as needed

# Step 3: Convert BGR to RGB format (for matplotlib)
mean_shifted_image_rgb = cv2.cvtColor(mean_shifted_image, cv2.COLOR_BGR2RGB)

# Step 4: Plot the segmented image
plt.figure(figsize=(4 ,5))
plt.imshow(mean_shifted_image_rgb)
plt.title('Mean-Shift Segmented Image')
plt.axis('off')
plt.show()
```

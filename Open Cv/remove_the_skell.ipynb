```python
import nibabel as nib
from nilearn.image import resample_to_img
from nilearn.masking import compute_brain_mask

# Load the MRI image
mri_img = nib.load('CT_Abdo.nii')

# Compute the brain mask
brain_mask = compute_brain_mask(mri_img)

# Apply the mask to the MRI image
brain_img = mri_img.get_fdata() * brain_mask.get_fdata()

# Save the result
brain_img_nifti = nib.Nifti1Image(brain_img, mri_img.affine)
nib.save(brain_img_nifti, 'output_brain_image.nii.gz')
```

```python
import nibabel as nib
import matplotlib.pyplot as plt

# Load a NIfTI file
nifti_file = nib.load('CT_Abdo.nii')

# Get image data
image_data = nifti_file.get_fdata()

# Plotting a slice of the image (for example, the middle slice along the z-axis)
z_slice = image_data.shape[-1] // 2  # Choose the middle slice along z-axis
plt.imshow(image_data[:, :, z_slice].T, cmap='gray', origin='lower')
plt.axis('off')  # Turn off axis labels
plt.title('MRI Slice')
plt.colorbar()
plt.show()
```

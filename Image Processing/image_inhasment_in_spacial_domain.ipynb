Log Transformation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Load the image
image = cv2.imread('2.jpg', cv2.IMREAD_GRAYSCALE)
resize=cv2.resize(image,(400,400))
# Apply log transformation
c = 255 / np.log(1 + np.max(resize))
log_transformed = c * np.log(1 + resize)
log_transformed = np.array(log_transformed, dtype=np.uint8)

plot_the_image = plt.imshow(log_transformed)
plt.show()

# Save and display the result
cv2.imwrite('log_transformed_2.jpg', log_transformed)
cv2.imshow('Log Transformed', log_transformed)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Power Law (Gama transformation)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('2.jpg', cv2.IMREAD_GRAYSCALE)

# Resize the image (optional)
resize = cv2.resize(image, (400, 400))

# Apply gamma correction
gamma = 2.0
gamma_corrected = np.array(255 * (resize / 255) ** gamma, dtype='uint8')

# Plot original and gamma-corrected images side-by-side
plt.subplot(1, 2, 1)  # Create subplot 1 in a row of 2
plt.imshow(image, cmap='gray')  # Plot original image in grayscale
plt.title('Original Image')
plt.axis('off')  # Hide axis labels and ticks

plt.subplot(1, 2, 2)  # Create subplot 2 in the same row
plt.imshow(gamma_corrected, cmap='gray')  # Plot gamma-corrected image in grayscale
plt.title('Gamma Corrected (gamma=2.0)')
plt.axis('off')

plt.show()
```

Contrast Stretching

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('2.jpg', cv2.IMREAD_GRAYSCALE)

# Apply contrast stretching
min_val = np.min(image)
max_val = np.max(image)
contrast_stretched = 255 * (image - min_val) / (max_val - min_val)
contrast_stretched = np.array(contrast_stretched, dtype=np.uint8)

# Plot original and contrast-stretched images side-by-side
plt.subplot(1, 2, 1)  # Create subplot 1 in a row of 2
plt.imshow(image, cmap='gray')  # Plot original image in grayscale
plt.title('Original Image')
plt.axis('off')  # Hide axis labels and ticks

plt.subplot(1, 2, 2)  # Create subplot 2 in the same row
plt.imshow(contrast_stretched, cmap='gray')  # Plot contrast-stretched image in grayscale
plt.title('Contrast Stretched')
plt.axis('off')

plt.show()
```

Thresholing 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('2.jpg', cv2.IMREAD_GRAYSCALE)

# Apply binary thresholding
threshold_value = 127  # You can adjust this value based on your image
_, thresholded = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

# Plot original and thresholded images side-by-side
plt.subplot(1, 2, 1)  # Create subplot 1 in a row of 2
plt.imshow(image, cmap='gray')  # Plot original image in grayscale
plt.title('Original Image')
plt.axis('off')  # Hide axis labels and ticks

plt.subplot(1, 2, 2)  # Create subplot 2 in the same row
plt.imshow(thresholded, cmap='gray')  # Plot thresholded image in grayscale (black and white)
plt.title('Thresholded (threshold=' + str(threshold_value) + ')')
plt.axis('off')

plt.show()
```

 Sharpening Filters

Laplacian Filter:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('2.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Laplacian filter
laplacian_filtered = cv2.Laplacian(image, cv2.CV_64F)
laplacian_filtered = np.uint8(np.absolute(laplacian_filtered))  # Convert to uint8 for visualization

# Plot original and Laplacian-filtered images side-by-side
plt.subplot(1, 2, 1)  # Create subplot 1 in a row of 2
plt.imshow(image, cmap='gray')  # Plot original image in grayscale
plt.title('Original Image')
plt.axis('off')  # Hide axis labels and ticks

plt.subplot(1, 2, 2)  # Create subplot 2 in the same row
plt.imshow(laplacian_filtered, cmap='gray')  # Plot Laplacian-filtered image in grayscale
plt.title('Laplacian Filtered (Edge Detection)')
plt.axis('off')

plt.show()
```

Unsharp Masking:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('2.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Gaussian blur
kernel_size = (9, 9)
sigma = 10.0  # Adjust sigma for blur strength
blurred = cv2.GaussianBlur(image, kernel_size, sigma)

# Perform unsharp masking
weight = 1.5
amount = -0.5
unsharp_masked = cv2.addWeighted(image, weight, blurred, amount, 0)

# Plot original, blurred, and sharpened images
plt.subplot(1, 3, 1)  # Create subplot 1 in a row of 3
plt.imshow(image, cmap='gray')  # Plot original image in grayscale
plt.title('Original Image')
plt.axis('off')  # Hide axis labels and ticks

plt.subplot(1, 3, 2)  # Create subplot 2 in the same row
plt.imshow(blurred, cmap='gray')  # Plot blurred image in grayscale
plt.title('Gaussian Blurred (Kernel Size: ' + str(kernel_size) + ', Sigma: ' + str(sigma) + ')')
plt.axis('off')

plt.subplot(1, 3, 3)  # Create subplot 3 in the same row
plt.imshow(unsharp_masked, cmap='gray')  # Plot unsharp-masked image in grayscale
plt.title('Unsharp Masked (Weight: ' + str(weight) + ', Amount: ' + str(amount) + ')')
plt.axis('off')

plt.show()
```

3. Histogram-Based Techniques

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('2.jpg', cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization
equalized = cv2.equalizeHist(image)

# Plot original and equalized images side-by-side
plt.subplot(1, 2, 1)  # Create subplot 1 in a row of 2
plt.imshow(image, cmap='gray')  # Plot original image in grayscale
plt.title('Original Image')
plt.axis('off')  # Hide axis labels and ticks

plt.subplot(1, 2, 2)  # Create subplot 2 in the same row
plt.imshow(equalized, cmap='gray')  # Plot equalized image in grayscale
plt.title('Histogram Equalized')
plt.axis('off')

plt.show()
```

 Histogram Matching (Specification)

```python
import cv2
import numpy as np
from skimage.exposure import match_histograms
import matplotlib.pyplot as plt

# Load the images
image = cv2.imread('2.jpg', cv2.IMREAD_GRAYSCALE)
reference = cv2.imread('log_transformed_2.jpg', cv2.IMREAD_GRAYSCALE)

# Apply histogram matching (remove multichannel argument)
matched = match_histograms(image, reference)

# Plot original, reference, and matched images side-by-side
plt.subplot(1, 3, 1)  # Create subplot 1 in a row of 3
plt.imshow(image, cmap='gray')  # Plot original image in grayscale
plt.title('Original Image')
plt.axis('off')  # Hide axis labels and ticks

plt.subplot(1, 3, 2)  # Create subplot 2 in the same row
plt.imshow(reference, cmap='gray')  # Plot reference image in grayscale
plt.title('Reference Image')
plt.axis('off')

plt.subplot(1, 3, 3)  # Create subplot 3 in the same row
plt.imshow(matched, cmap='gray')  # Plot matched image in grayscale
plt.title('Histogram Matched')
plt.axis('off')

plt.show()
```
